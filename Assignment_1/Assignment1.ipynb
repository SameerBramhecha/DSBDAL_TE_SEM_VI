## DEALING WITH MISSING VALUES

```python
df.isna().sum()
```

```python
missing  =df[df.isnull().any(axis=1)]
missing
```

```python
df.describe()
```

```python
round(df['Age'].mean())
```

```python
x = round(df['Age'].mean())
df['Age'].fillna(x,inplace=True)
df['Age']
```

```python
df['KM'].median()
```

```python
df['KM'].fillna(df['KM'].median(),inplace=True)
df['KM']
```

```python
df['HP'].mean()
```

```python
df['HP'].fillna(round(df['HP'].mean()),inplace=True)
df['HP']
```

```python
df.isna().sum()
```

```python
df['FuelType'].value_counts()
```

```python
df['FuelType'].fillna(df['FuelType'].value_counts().index[0],inplace=True)
df['FuelType']
```

```python
df['MetColor'].mode()
```

```python
df['MetColor'].fillna(df['MetColor'].mode()[0],inplace=True)
```
